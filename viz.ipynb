```python
import pandas as pd
import pandas_bokeh
import matplotlib.pyplot as plt

pandas_bokeh.output_notebook()

tweets = pd.read_csv("data.csv", parse_dates=["timestamp"])

# The account tweets twice a day, at roughly 4am and 4pm. Filtering out the
# morning tweets makes the signal less noisy.
tweets = tweets[tweets.timestamp.dt.hour >= 12]

indexed_tweets = tweets.set_index("timestamp")
indexed_tweets["likes"].plot_bokeh()
```
